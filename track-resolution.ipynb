```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
B = 3.8 # V*s/m^2
qB = B*(3*10**8) # eV/m
```

```python
def deltaPhi(pT,R):
    pT = pT*10**9 # eV
    return np.arcsin(R*qB/(2*pT))
```

```python
xmin = 0.5
xmax = 2*np.pi - 0.1

x = np.linspace(xmin,xmax,100)
y1 = deltaPhi(x,0.05)*180/(np.pi)
y2 = deltaPhi(x,0.1)*180/(np.pi)
y3 = deltaPhi(x,0.2)*180/(np.pi)

plt.plot(x,y1,label='R=50mm',color='blue')
plt.plot(x,y2,label='R=100mm',color='red')
plt.plot(x,y3,label='R=200mm',color='green')
plt.legend()
plt.xlabel('$p_T$ [GeV]')
plt.ylabel(r'$\Delta\phi$ [deg]')
```

```python
def cotBeta(pT,R):
    return 1./np.tan(deltaPhi(pT,R))
```

```python
xmin = 0.1
xmax = 2*np.pi - 0.1
R = 0.01 # innermost pixel layer = 4 cm

x = np.linspace(xmin,xmax,100)
y1 = 1./np.tan(x)

plt.plot(x,y1,label='R=40mm',color='blue')
plt.legend()
plt.xlabel('$x$ [GeV]')
plt.ylabel(r'cot($x$)')
```

```python
xmin = 0.5
xmax = 3

x = np.linspace(xmin,xmax,100)
y1 = cotBeta(x,0.05)
y2 = cotBeta(x,0.1)
y3 = cotBeta(x,0.2)

plt.plot(x,y1,label='R=50mm',color='blue')
plt.plot(x,y2,label='R=100mm',color='red')
plt.plot(x,y3,label='R=200mm',color='green')
plt.legend()
plt.xlabel('$p_T$ [GeV]')
plt.ylabel(r'$\cot\beta$')
```

```python
def pT(cotBeta,R):
    dPhi = np.arctan(1./cotBeta)
    return R*qB/(2*np.sin(dPhi))/(10**9) # eV
```

```python
# Sanity check
pt = 3
R = .5
pT(cotBeta(pt,R),R)
```

```python
xmin = 0
xmax = 30

x = np.linspace(xmin,xmax,100)
y1 = pT(x,0.05)
y2 = pT(x,0.1)
y3 = pT(x,0.2)

plt.plot(x,y1,label='R=50mm',color='blue')
plt.plot(x,y2,label='R=100mm',color='red')
plt.plot(x,y3,label='R=200mm',color='green')
plt.legend()
plt.ylabel('$p_T$ [GeV]')
plt.xlabel(r'$\cot\beta$')
```

```python
xmin = 0.1
xmax = 10

R = 0.05
x = np.linspace(xmin,xmax,100)
y1 = pT(x,R)

y2 = pT(x+0.01*np.ones(100),R)/y1
y3 = pT(x-0.01*np.ones(100),R)/y1

y4 = pT(x+0.1*np.ones(100),R)/y1
y5 = pT(x-0.1*np.ones(100),R)/y1


plt.plot(x,y2,label=r'$\sigma = +0.01$',color='blue')
plt.plot(x,y3,label=r'$\sigma = -0.01$',color='blue',linestyle='--')
plt.plot(x,y4,label=r'$\sigma = +0.1$',color='red')
plt.plot(x,y5,label=r'$\sigma = -0.1$',color='red',linestyle='--')

plt.legend()
plt.ylabel(r'$p_T (\cot\beta \pm \sigma) / p_T (\cot\beta)$')
plt.xlabel(r'$\cot\beta$')
```

```python
xmin = 0.1
xmax = 20

R = 0.5
x = np.linspace(xmin,xmax,100)
y1 = pT(x,R)

y2 = np.divide(pT(x+0.01,R),y1)
y3 = np.divide(pT(x-0.01,R),y1)

y4 = np.divide(pT(x+0.1,R),y1)
y5 = np.divide(pT(x-0.1,R),y1)

plt.plot(y1,y2,label=r'$\sigma = +0.01$',color='blue')
plt.plot(y1,y3,label=r'$\sigma = -0.01$',color='blue',linestyle='--')
plt.plot(y1,y4,label=r'$\sigma = +0.1$',color='red')
plt.plot(y1,y5,label=r'$\sigma = -0.1$',color='red',linestyle='--')

plt.legend()
plt.ylabel(r'$p_T (\cot\beta \pm \sigma) / p_T (\cot\beta)$')
plt.xlabel(r'$p_T (\cot\beta)$')
```

```python
xmin = 0.1
xmax = 20

R = 0.5
x = np.linspace(xmin,xmax,100)
y1 = pT(x,R)

y2 = pT(x+0.01,R) - y1
y3 = pT(x-0.01,R) - y1

y4 = pT(x+0.1,R) - y1
y5 = pT(x-0.1,R) - y1

plt.plot(y1,y2,label=r'$\sigma = +0.01$',color='blue')
plt.plot(y1,y3,label=r'$\sigma = -0.01$',color='blue',linestyle='--')
plt.plot(y1,y4,label=r'$\sigma = +0.1$',color='red')
plt.plot(y1,y5,label=r'$\sigma = -0.1$',color='red',linestyle='--')

plt.legend()
plt.ylabel(r'$p_T (\cot\beta \pm \sigma) - p_T (\cot\beta)$')
plt.xlabel(r'$p_T (\cot\beta)$')
```
