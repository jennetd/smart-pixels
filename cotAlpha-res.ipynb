```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cot(x):
    return 1./np.tan(x)
    
def arccot(x):
    return np.arctan(1./x)
    
def deltaZ(theta,R):
    return R/np.tan(theta)

def Eta(theta):
    return -np.log(np.tan(theta/2))

def Theta(eta):
    return 2*np.arctan(np.exp(-1*eta))
```

```python
Theta(Eta(2))
```

```python
xmin = -np.pi/2
xmax = np.pi/2

x = np.linspace(xmin,xmax,100)
y1 = deltaZ(x,0.05)
y2 = deltaZ(x,0.1)
y3 = deltaZ(x,0.2)

plt.plot(x*180/(np.pi),y1,label='R=50mm',color='blue')
plt.plot(x*180/(np.pi),y2,label='R=100mm',color='red')
plt.plot(x*180/(np.pi),y3,label='R=200mm',color='green')
plt.legend()
plt.xlabel(r'$\alpha$ [deg]')
plt.ylabel(r'$\Delta z$ [mm]')
```

```python
xmin = -5
xmax = 5

x = np.linspace(xmin,xmax,100)
y1 = deltaZ(Theta(x),0.05)#*180/(np.pi)
y2 = deltaZ(Theta(x),0.1)#*180/(np.pi)
y3 = deltaZ(Theta(x),0.2)#*180/(np.pi)

plt.plot(x,y1,label='R=50mm',color='blue')
plt.plot(x,y2,label='R=100mm',color='red')
plt.plot(x,y3,label='R=200mm',color='green')
plt.legend()
plt.xlabel(r'$\eta$')
plt.ylabel(r'$\Delta z$ [mm]')
```

```python
xmin = -np.pi/2
xmax = np.pi/2

x = np.linspace(xmin,xmax,100)
y1 = deltaZ(x,0.05)
y2 = deltaZ(x,0.1)
y3 = deltaZ(x,0.2)

plt.plot(cot(x),y1,label='R=50mm',color='blue')
plt.plot(cot(x),y2,label='R=100mm',color='red')
plt.plot(cot(x),y3,label='R=200mm',color='green')
plt.legend()
plt.xlabel(r'$\cot \alpha$')
plt.ylabel(r'$\Delta z$ [mm]')
```

```python
xmin = -0.2
xmax = 0.2

res = 0.01
x = np.linspace(xmin,xmax,100)

R = 0.05
y1 = 1000*deltaZ(np.arctan(1/x),R)
plt.plot(y1,x,label='R=50mm',color='blue')
plt.plot([1000*deltaZ(arccot(res),R),1000*deltaZ(arccot(res),R)],[-100,res],color='blue')
plt.plot([1000*deltaZ(arccot(-1*res),R),1000*deltaZ(arccot(-1*res),R)],[-100,-1*res],color='blue')

R = 0.10
y2 = 1000*deltaZ(np.arctan(1/x),R)
plt.plot(y2,x,label='R=100mm',color='red')
plt.plot([1000*deltaZ(arccot(res),R),1000*deltaZ(arccot(res),R)],[-100,res],color='red')
plt.plot([1000*deltaZ(arccot(-1*res),R),1000*deltaZ(arccot(-1*res),R)],[-100,-1*res],color='red')

plt.legend(loc='upper left')
plt.ylabel(r'$\cot \alpha$')
plt.xlabel(r'$\Delta z$ [um]')
plt.ylim(xmin,xmax)
```
