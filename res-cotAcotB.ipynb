```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
true = pd.read_csv('/eos/uscms/store/user/jennetd/smart-pixels/from-jieun/cotAcotB-baseline/trueLabels.csv',index_col=0)
pred = pd.read_csv('/eos/uscms/store/user/jennetd/smart-pixels/from-jieun/cotAcotB-baseline/predictions.csv',index_col=0)
```

```python
true['cotBeta_res'] = true['cotBeta']-pred['cotBeta']
true['cotAlpha_res'] = true['cotAlpha']-pred['cotAlpha']
```

```python
plt.hist(true['cotBeta_res'],bins = 100);
plt.xlabel(r'True - pred $\cot \beta$')
```

```python
plt.hist2d(x=true['cotBeta'],y=true['cotBeta_res'],bins=(50,50));
plt.xlabel(r'True $\cot \beta$')
plt.ylabel(r'True - Pred $\cot \beta$')
plt.colorbar()
```

```python
plt.hist(true['cotAlpha_res'],bins = 100);
plt.xlabel(r'True - pred $\cot \alpha$')
```

```python
plt.hist2d(x=true['cotAlpha'],y=true['cotAlpha_res'],bins=(50,50));
plt.xlabel(r'True $\cot \alpha$')
plt.ylabel(r'True - Pred $\cot \alpha$')
plt.colorbar()
```
